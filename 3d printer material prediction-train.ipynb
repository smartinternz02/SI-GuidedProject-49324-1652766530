# Missing value treatment

```python
data1.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
mean_rou=data1['roughness'].mean()
mean_ten=data1['tension_strenght'].mean()
mean_elo=data1['elongation'].mean()
```

```python
data1 ['roughness'].fillna(mean_rou, inplace=True)
data1 ['tension_strenght'].fillna(mean_ten, inplace=True)
data1 ['elongation'].fillna(mean_elo, inplace=True)
```

```python
data1.apply(lambda x: sum(x.isnull()),axis=0)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data1['infill_pattern']=lb.fit_transform(data1['infill_pattern'])
data1['material']=lb.fit_transform(data1['material'])
```

```python
data1
```

# One Hot Encoding

```python
data1.dtypes
```

```python
data1['infill_pattern'].value_counts()
```

```python
data1['material'].value_counts()
```

```python
data1 = pd.get_dummies(data1)
data1.head()
```

```python
data1
```

```python
x=data1.loc[:, ['roughness','tension_strenght','elongation']]
y=data1.iloc[:,-1]
```

```python
x
```

```python
y
```

# Feature Scaling

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

# Splittimg data into Train and Test

```python
#Splitting the dataset into Training set and Test set
from sklearn import model_selection, neighbors
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train = y_train.astype('int')
y_train
```

```python
y_test = y_test.astype('int')
y_test
```

# Training and Testing the Model

```python
#Fitting Classifier to the Training set
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
#Predicting the Test Set results
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#Accuracy score on Test and Train
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix

print("\nAccuracy score: %f" %(accuracy_score(y_test,y_pred) * 100))
print("Recall score : %f" %(recall_score(y_test, y_pred) * 100))
print("ROC score : %f\n" %(roc_auc_score(y_test, y_pred) * 100))
print(confusion_matrix(y_test, y_pred)) 
```

```python
#Predict the probability of each test data point
#Get the probability distribution
probas = clf.predict_proba(x_test)
```

```python
#Visualize the probabilities
# plot
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.hist(probas, bins=20)
plt.title('Classification Probabilities')
plt.xlabel('Probability')
plt.ylabel('# of Instances')
plt.xlim([0.5, 1.0])
plt.legend(y_test)
plt.show()
```

# Evaluation

```python
x_train_std = minmax.fit_transform(x_train)
x_test_std = minmax.transform(x_test)
```

```python
#Evaluation Metrics
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
#Print cross validation score for kNN
clf_acc = cross_val_score(clf, x_train_std, y_train, cv=3, scoring="accuracy",  n_jobs=-1)
```
